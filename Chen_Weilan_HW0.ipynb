# Pandas

```python
#Pandas abc
df = pd.read_csv("../data/Salaries.csv",header=0,index_col="playerID",skiprows=[1])
df
```

```python
#Pandas d
df[(df['teamID'].isin(["ATL","HOU"])) & (df['salary']>1000000)]
```

```python
#Pandas e
df[df['teamID']=='ATL'].describe()
```

```python
#Pandas f
dic={}
dic["playerID"]=[]
for i in df.columns.values:
    dic[i]=[]
it=df.iterrows()
for i in it:
    dic["playerID"].append(i[0])
    for (j,k) in enumerate(df.columns.values):
        dic[k].append(i[1][j])
        
dic
```

```python
#Pandas g
df2=pd.DataFrame.from_dict(dic)
df2.columns=['a','b','c','d','e']
df2
```

```python
# Numpy a
li = [[1,2],[3,4]]
arr = np.array(li)
arr
```

```python
# Numpy b
print(arr.ndim)
print(arr.shape)
print(arr.size)
print(arr.dtype)
print(arr.itemsize)
print(arr.data)
```

```python
# Numpy c
arr = arr.flatten()
print(arr)
arr = arr.reshape([2,2])
print(arr)
```

```python
# Numpy d
arr[:,::-1]
```

```python
# Numpy e
print(arr.argmin(axis=1))
print(arr.argmax())
print(arr.min())
print(arr.max())
print(arr.mean())
print(arr.sum())
print(arr.std())
print(np.dot(arr,arr))
print(arr**2)
print(arr**0.5)
print(np.abs(arr))
print(np.exp(arr))
print(np.sign(arr))
print(np.mod(arr,2))
```
